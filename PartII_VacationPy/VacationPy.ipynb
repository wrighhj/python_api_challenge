```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_data = "../output_data/weather_data.csv"
weather = pd.read_csv(weather_data)
weather
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = weather[["Latitude", "Longitude"]].astype(float)

# Convert Humidity to float and store
weather = weather.dropna()
humidity = weather["Humidity"].astype(float)
```

```python
# Create a poverty Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)

fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
good_weather = weather.loc[(weather["Temperature"] > 70) & (weather["Temperature"] < 80) & (weather["Wind Speed"] < 10) & (weather["Cloudiness"] == 0), :]
good_weather = good_weather.dropna(how='any')
good_weather.reset_index(inplace=True)
del good_weather['index']
good_weather.head()
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotels = []

hotel_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

for i in range(len(good_weather)):

    lat = good_weather.loc[i]['Latitude']
    lon = good_weather.loc[i]['Longitude']
    
    params = {"location": f"{lat}, {lon}",
        "radius": 5000,
        "types" : "hotel",
        "key": g_key}
    
    response = requests.get(hotel_url, params=params).json()
    
    try:
        hotels.append(response['results'][0]['name'])
    except:
        hotels.append("")
        
good_weather["Hotel"] = hotels
good_weather = good_weather.dropna(how='any')
good_weather
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in good_weather.iterrows()]
locations = good_weather[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)

# Display figure
fig
```
